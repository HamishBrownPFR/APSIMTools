## Opening data from APSIM

```python
con = sqlite3.connect(r'C:\Users\jjojeda\Dropbox\DATA\APSIM sim\PaperI.db')
```

```python
Harvest = pd.read_sql("Select * from HarvestReport",
                        con)
```

```python
Harvest.columns
```

## Creating variables to extract weather and position data

```python
station=Harvest.loc[:,'Weather.FileName']
lat=Harvest.groupby(station)[["Weather.Latitude"]].mean()
long=Harvest.groupby(station)[["Weather.Longitude"]].mean()
tav=Harvest.groupby(station)[["Weather.Tav"]].mean()
coo=pd.concat([lat, long, tav], axis=1)
```

```python
coo.columns
```

```python
coo
```

```python
longs = pd.Series(coo['Weather.Longitude'].values)
lats = pd.Series(coo['Weather.Latitude'].values)
tav = pd.Series(coo['Weather.Tav'].values)
lat1= lats.tolist()
lon1= longs.tolist()
tav1= tav.tolist()
```

```python
tav1
```
