```python
# Defining variables's names
maxt=Daily.loc[:,'Weather.MaxT']
mint=Daily.loc[:,'Weather.MinT']
rain=Daily.loc[:,'Weather.Rain']
rad=Daily.loc[:,'Weather.Radn']
jd=Daily.loc[:,'Clock.Today.DayOfYear']
# Calculating means of each variable between years
meanjd=Daily.groupby(jd)[["Clock.Today.DayOfYear"]].mean()
meanmaxt=Daily.groupby(jd)[["Weather.MaxT"]].mean()
sdmaxt=Daily.groupby(jd)[["Weather.MaxT"]].std()
meanmint=Daily.groupby(jd)[["Weather.MinT"]].mean()
sdmint=Daily.groupby(jd)[["Weather.MinT"]].std()
meanradn=Daily.groupby(jd)[["Weather.Radn"]].mean()
sdradn=Daily.groupby(jd)[["Weather.Radn"]].std()
summonthlyrain=Daily.groupby([' Soil', '  Water','  P_Weather','Clock.Today.Year','Clock.Today.Month'])[["Weather.Rain"]].sum()
meanmonthlyrain=summonthlyrain.groupby('Clock.Today.Month')[["Weather.Rain"]].mean()
stdmonthlyrain=summonthlyrain.groupby('Clock.Today.Month')[["Weather.Rain"]].std()
# Converting from DataFrame 2D to numpy array 1D
jd1d=np.array(meanjd['Clock.Today.DayOfYear'])
meanmaxt1d=np.array(meanmaxt['Weather.MaxT'])
sdmaxt1d=np.array(sdmaxt['Weather.MaxT'])
meanmint1d=np.array(meanmint['Weather.MinT'])
sdmint1d=np.array(sdmint['Weather.MinT'])
meanrad1d=np.array(meanradn['Weather.Radn'])
sdrad1d=np.array(sdradn['Weather.Radn'])
meanrain1d=np.array(meanmonthlyrain['Weather.Rain'])
stdrain1d=np.array(stdmonthlyrain['Weather.Rain'])
```

```python
# set up figure and ax
fig, ax = plt.subplots(figsize=(8,5))
# create the scatter plot
plt.plot(jd,maxt,".", color=(0.5,0.5,0.5), markersize=1)
plt.xlabel('')
plt.ylabel('MaxT (°C)', fontsize =14)
plt.xlim(-20,386)
plt.ylim(-5,45)
plt.tick_params(direction='in', length=6, width=1, colors='k',
               grid_color='k', grid_alpha=0.5, bottom=bool, top=bool, 
                left=bool, right=bool)
plt.rcParams['figure.figsize']=(10,5)
ax = plt.gca()
ax.grid(True, 'major', color='w', linestyle='-', linewidth=1.4)
ax.grid(True, 'minor', color='0.92', linestyle='-', linewidth=0.7)
ax.patch.set_facecolor('0.85')
ax.set_axisbelow(True)
# create the line
ax.plot(meanjd,meanmaxt, '-',linewidth=1, color='k')
```
