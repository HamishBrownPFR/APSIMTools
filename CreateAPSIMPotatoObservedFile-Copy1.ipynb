```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Bring biomass data into dataframe

```python
Obs = pd.read_excel(r'C:\GitHubRepos\ApsimX\Prototypes\Potato\Observed.xlsx')
```

```python
for o in Obs.index:
    if np.isnan(Obs.loc[o,'Potato.AboveGround.Wt']):
        Obs.loc[o,'Potato.AboveGround.Wt'] = Obs.loc[o,'Potato.AboveGround.Wt.1']
Obs.drop('Potato.AboveGround.Wt.1',axis=1,inplace=True)
```

```python
for o in Obs.index:
    if np.isnan(Obs.loc[o,'Potato.Leaf.AppearedCohortNo']):
        Obs.loc[o,'Potato.Leaf.AppearedCohortNo'] = Obs.loc[o,'Potato.Structure.LeafTipsAppeared']
Obs.drop('Potato.Structure.LeafTipsAppeared',axis=1,inplace=True)
```

```python
for o in Obs.index:
    if Obs.loc[o,'PlantandHarvest.Script.FinalTag'] != 'yield':
        Obs.loc[o,'PlantandHarvest.Script.FinalTag'] = Obs.loc[o,'PlantAndHarvest.Script.FinalTag']
Obs.drop('PlantAndHarvest.Script.FinalTag',axis=1,inplace=True)
```

```python
for o in Obs.index:
    if np.isnan(Obs.loc[o,'Potato.AboveGround.Live.N']):
        Obs.loc[o,'Potato.AboveGround.Live.N'] = Obs.loc[o,'Potato.AboveGroundLive.N']/100
Obs.drop('Potato.AboveGroundLive.N',axis=1,inplace=True)
```

```python
for o in Obs.index:
    if np.isnan(Obs.loc[o,'TotalNO3']):
        Obs.loc[o,'TotalNO3'] = Obs.loc[o,'TotalNO3.1']
Obs.drop('TotalNO3.1',axis=1,inplace=True)
```

```python
for o in Obs.index:
    if np.isnan(Obs.loc[o,'TotalNH4']):
        Obs.loc[o,'TotalNH4'] = Obs.loc[o,'TotalNH4.1']
Obs.drop('TotalNH4.1',axis=1,inplace=True)
```

```python
for o in Obs.index:
    if np.isnan(Obs.loc[o,'Potato.TotalLive.N']):
        Obs.loc[o,'Potato.TotalLive.N'] = Obs.loc[o,'Potato.Total.N']
Obs.drop('Potato.Total.N',axis=1,inplace=True)
```

```python
Obs.loc[:,'Potato.AboveGround.Live.Nconc'] = Obs.loc[:,'Potato.AboveGroundLive.NConc']/100
Obs.drop('Potato.AboveGroundLive.NConc',axis=1,inplace=True)
```

```python
Obs.loc[:,'Potato.AboveGround.Live.Wt'] = Obs.loc[:,'Potato.AboveGround.Wt']
Obs.drop('Potato.AboveGround.Wt',axis=1,inplace=True)
```

```python
#Drop out variables that are no use to reduce clutter
Obs.drop(['Experiment',
 'ExtraVariables.Script.AccumTtemerg',
 'LeafNodeNoerror',
 'Potato.AboveGroundLive.FWt',
 'Potato.Leaf.Live.FWt',
 'Potato.Phenology.CurrentPhaseName',
 'Potato.Stem.Live.FWt',
 'TDWA',
 'TWAD',
 'cultivar',
 'Soil.SoilNitrogen.NH4ppm(1)',
 'Soil.SoilNitrogen.NH4ppm(2)',
 'Soil.SoilNitrogen.NH4ppm(3)',
 'Soil.SoilNitrogen.NH4ppm(4)',
 'Soil.SoilNitrogen.NO3ppm(1)',
 'Soil.SoilNitrogen.NO3ppm(2)',
 'Soil.SoilNitrogen.NO3ppm(3)',
 'Soil.SoilNitrogen.NO3ppm(4)',],axis=1,inplace=True)
```

```python
Obs.dropna(axis=1,how='all',inplace=True)
```

```python
#Some tuber N conc were in the wrong units, convert
for o in Obs.index:
    if Obs.loc[o,'Potato.Tuber.Live.NConc'] > 0.3:
        Obs.loc[o,'Potato.Tuber.Live.NConc']= Obs.loc[o,'Potato.Tuber.Live.NConc'] /100
```

```python
# Some leaf Nconc obs were claerly wrong.  Remove
for o in Obs.index:
    if Obs.loc[o,'Potato.Leaf.Live.NConc'] > 1:
        Obs.loc[o,'Potato.Leaf.Live.NConc']= np.nan
```

```python
for o in Obs.index:
    if np.isnan(Obs.loc[o,'Potato.Leaf.Live.N']):
        Obs.loc[o,'Potato.Leaf.Live.N'] = Obs.loc[o,'Potato.Leaf.Live.Wt'] * Obs.loc[o,'Potato.Leaf.Live.NConc']
for o in Obs.index:
    if np.isnan(Obs.loc[o,'Potato.Tuber.Live.N']):
        Obs.loc[o,'Potato.Tuber.Live.N'] = Obs.loc[o,'Potato.Tuber.Live.Wt'] * Obs.loc[o,'Potato.Tuber.Live.NConc']
for o in Obs.index:
    if np.isnan(Obs.loc[o,'Potato.Stem.Live.N']):
        Obs.loc[o,'Potato.Stem.Live.N'] = Obs.loc[o,'Potato.Stem.Live.Wt'] * Obs.loc[o,'Potato.Stem.Live.NConc']
for o in Obs.index:
    if np.isnan(Obs.loc[o,'Potato.TotalLive.N']):
        Obs.loc[o,'Potato.TotalLive.N'] = Obs.loc[o,'Potato.Tuber.Live.N'] +Obs.loc[o,'Potato.Leaf.Live.N'] + Obs.loc[o,'Potato.Stem.Live.N']
```

## Set indexes in the smaller tables and deal to duplicate rows

```python
from collections import Counter
def MakeMean(ser):
    Series = pd.Series(index=ser.columns)
    for Group in ser.columns:
        if isinstance(Group, float):
            Series[Group] = ser[Group].mean()
        else:
            Series[Group] = Counter(ser[Group]).most_common(1)[0][0]
    return Series        
```

```python
Obs.loc[:,'Clock.Today'] = pd.to_datetime(Obs.loc[:,'Clock.Today'])
Obs.set_index(['SimulationName','Clock.Today'],inplace=True)
Obs = Obs.groupby(level=['SimulationName','Clock.Today']).mean()
#Compressed = Compressed.groupby(level=['SimulationName','Clock.Today']).mean()
```

```python
Obs.sort_index(axis=0,inplace=True)
Obs.sort_index(axis=1,inplace=True)
```
