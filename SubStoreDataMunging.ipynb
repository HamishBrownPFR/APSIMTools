```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

## Make a list of observed files

```python
SubStoreFilePaths = []
mydir = r'C:\Users\cflhxb\Dropbox\APSIMPotato\Observed\SubStorValidationData\experiments'
for file in os.listdir(mydir):
    if file.endswith('.xlsx'):
        SubStoreFilePaths.append(os.path.join(mydir, file))
```

## Make a list of data frames to work through

```python
SubStoreFiles = []
for ObsTable in SubStoreFilePaths:
    SubStoreFiles.append(pd.read_excel(ObsTable,na_values=-99))
```

```python
def IsMetaDataRow(RowVal):
    if isinstance(RowVal, str):
        if 'Metadata' in RowVal:
            return True
        else:
            return False
    else: 
        return False

def IsTimeSeriesRow(RowVal):
    if isinstance(RowVal, str):
        if 'Time' in RowVal:
            return True
        else:
            return False
    else: 
        return False
    
def IsSummaryRow(RowVal):
    if isinstance(RowVal, str):
        if 'Summary' in RowVal:
            return True
        else:
            return False
    else:
        return False
    
        
def IsManagementRow(RowVal):
    if isinstance(RowVal, str):
        if 'Management' in RowVal:
            return True
        else:
            return False
    else:
        return False
```

## Go through each data file and put the experiment name and treatments into a index frame

```python
TreatList = []
for ObsTable in SubStoreFiles:
    MetaDataMask = [IsMetaDataRow(x) for x in ObsTable.iloc[:,0]]
    FirstTreatRowPos = ObsTable.iloc[MetaDataMask].index[0] + 5
    TreatRow = FirstTreatRowPos
    while isinstance(ObsTable.iloc[TreatRow,0], int):
        TreatNumber = ObsTable.iloc[TreatRow,0]
        ExpName = str(ObsTable.iloc[TreatRow][3] )
        TreatName = ExpName + str(ObsTable.iloc[TreatRow][5])
        TreatName = TreatName.replace(' ','')
        TreatName = TreatName.replace('=','')
        TreatName = TreatName.replace('%','')
        TreatList.append([ExpName,TreatNumber,TreatName])
        TreatRow +=1
TreatmentIndex = pd.DataFrame(data=TreatList, 
                          columns = ['ExpName','TreatNumber','TreatName'])
TreatmentIndex.set_index('TreatName',inplace=True)    
```

## Find the column headers for observed data and and add them into a column index

```python
ColumnHeaders = ['SimulationName','Clock.Today','TreatmentNumber']
for ObsTable in SubStoreFiles:
    TimeSeriesMask = [IsTimeSeriesRow(x) for x in ObsTable.iloc[:,0]]
    ColumnIndexRow = ObsTable.iloc[TimeSeriesMask].index[0] + 3
    ObsTable.iloc[ColumnIndexRow,0] = u'TreatmentNumber'
    ObsTable.iloc[ColumnIndexRow,1] = u'Clock.Today'
    member = 0
    for x in ObsTable.iloc[ColumnIndexRow].values:
        if isinstance(x, str):
            ObsTable.iloc[ColumnIndexRow,member] = x.replace('!','')
        member +=1
    member = 0
    for x in ObsTable.iloc[ColumnIndexRow].values:
        if isinstance(x, str):
            ObsTable.iloc[ColumnIndexRow,member] = x.replace('Ad','AD')
        member +=1
    member = 0
    for x in ObsTable.iloc[ColumnIndexRow].values:
        if isinstance(x, str):
            ObsTable.iloc[ColumnIndexRow,member] = x.replace('TUNA','PNAD')
        member +=1
    member = 0
    for x in ObsTable.iloc[ColumnIndexRow].values:
        if isinstance(x, str):
            ObsTable.iloc[ColumnIndexRow,member] = x.replace('UN%H','UN%D')
        member +=1
    member = 0
    for x in ObsTable.iloc[ColumnIndexRow].values:
        if isinstance(x, str):
            ObsTable.iloc[ColumnIndexRow,member] = x.replace('UYAFH','UYAD')
        member +=1
    member = 0
    for x in ObsTable.iloc[ColumnIndexRow].values:
        if isinstance(x, str):
            ObsTable.iloc[ColumnIndexRow,member] = x.replace('UYFAD','UYAD')
        member +=1
    
    ObsList = ObsTable.iloc[ColumnIndexRow,1:].dropna().values
    ObsList = [str(ObsList[x]) for x in range(len(ObsList))]
    #ObsList = [ObsList[x].replace('!','') for x in range(len(ObsList))]
    
    ColumnHeaders = list(set().union(ColumnHeaders, ObsList))
    
```

## Step through each datatable and put the data into the merged table

```python
Count = 0
CleanTabs = []
for ObsTable in SubStoreFiles:
    #Make a data frame with the treatment indexs for this table
    MetaDataMask = [IsMetaDataRow(x) for x in ObsTable.iloc[:,0]]
    FirstTreatRowPos = ObsTable.iloc[MetaDataMask].index[0] + 5
    TreatRow = FirstTreatRowPos
    TreatNumbers = []
    TreatNames = []
    while isinstance(ObsTable.iloc[TreatRow,0], int):
        TreatName = str(ObsTable.iloc[TreatRow,3] ) + str(ObsTable.iloc[TreatRow,5])
        TreatName = TreatName.replace(' ','')
        TreatName = TreatName.replace('=','')
        TreatName = TreatName.replace('%','')
        TreatNames.append([TreatName])
        TreatNumbers.append(ObsTable.iloc[TreatRow,0])
        TreatRow +=1
    TreatIndex = pd.DataFrame(index = TreatNumbers,
                              data=TreatNames,
                              columns=['TreatmentName'])
    #Find the observed data for this table
    TimeSeriesMask = [IsTimeSeriesRow(x) for x in ObsTable.iloc[:,0]]
    ColumnIndexRow = ObsTable.iloc[TimeSeriesMask].index[0] + 3
    FirstObsRowPos = ObsTable.iloc[TimeSeriesMask].index[0] + 4
          
    #Make a Dataframe with the observed data for this experiment
    Columns = ObsTable.iloc[ColumnIndexRow].values
    CleanTable = pd.DataFrame(data = ObsTable.iloc[FirstObsRowPos:].values,
                             columns = Columns)
    CleanTable.dropna(axis = 0, how='all',inplace=True)
    CleanTable.dropna(axis = 1, how='all',inplace=True)
    CleanTable.ix[:,'SimulationName'] = [TreatIndex.ix[x,'TreatmentName'] for x in CleanTable.ix[:,'TreatmentNumber'].values]
    CleanTabs.append(CleanTable)
    Headers = pd.DataFrame(data = ObsTable.iloc[FirstObsRowPos-3:FirstObsRowPos].transpose().values,
                          columns = ['Definitions','Units','Code'] )
    HeaderInfo = HeaderInfo.append(Headers,ignore_index=True)
    Count +=1
    
```
